```python
import os

root_dir = "root" # Path to your dataset
image_extensions = {'.jpg', '.jpeg', '.png'}

counter = 1

for split in ["train", "test", "valid"]:
    img_dir = os.path.join(root_dir, split, "images")
    lbl_dir = os.path.join(root_dir, split, "labels")

    for file in os.listdir(img_dir):
        ext = os.path.splitext(file)[1].lower()
        if ext not in image_extensions:
            continue
        
        img_old = os.path.join(img_dir, file)
        base = os.path.splitext(file)[0]
        lbl_old = os.path.join(lbl_dir, base + ".txt")

        if not os.path.exists(lbl_old):
            print(f"⚠️ Label missing for {file}, skipped")
            continue

        # Create new safe names
        new_name = f"sample_{counter:05d}"
        img_new_temp = os.path.join(img_dir, new_name + "_TMP" + ext)
        lbl_new_temp = os.path.join(lbl_dir, new_name + "_TMP.txt")

        # Rename to temp first (avoid conflicts)
        os.rename(img_old, img_new_temp)
        os.rename(lbl_old, lbl_new_temp)

        counter += 1

# Rename _TMP files to final names
for split in ["train", "test", "valid"]:
    img_dir = os.path.join(root_dir, split, "images")
    lbl_dir = os.path.join(root_dir, split, "labels")

    for file in os.listdir(img_dir):
        if "_TMP" in file:
            new_final = file.replace("_TMP", "")
            os.rename(os.path.join(img_dir, file), os.path.join(img_dir, new_final))

    for file in os.listdir(lbl_dir):
        if "_TMP" in file:
            new_final = file.replace("_TMP", "")
            os.rename(os.path.join(lbl_dir, file), os.path.join(lbl_dir, new_final))

print("\n✅ Done renaming! No conflicts, all labels matched.")
print(f"Total renamed images: {counter - 1}")
```
